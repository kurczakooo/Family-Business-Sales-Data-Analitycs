### Here we make an effort to extract valuable info from the aggreggated data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style(style='darkgrid')
sns.set_context("paper")
sns.set_palette('dark')
```

```python
sales_df = pd.read_csv('../out/sales_aggregated.csv', index_col=0)
```

```python
sales_df.head()
```

```python
plt.figure(figsize=(24,4))
sns.lineplot(x=sales_df.columns, y=sales_df.iloc[0, :])
plt.title('SCHAB Z KOŚCIĄ sales')
plt.xlabel('Month')
plt.ylabel('Kilograms sold')
plt.xticks(rotation=45)
plt.show()
```
