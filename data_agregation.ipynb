### Here we aggregate and transform all the sales data and export it

```python
import os
import numpy as np
import pandas as pd
```

```python
data_files = os.listdir('./out/')
data_files
```

```python
csv_list = []
for file in data_files:
    df = pd.read_csv(os.path.join('./out/', file), index_col="Produkt")
    csv_list.append(df)
    
for df in csv_list:
   df.drop("Unnamed: 0", axis=1, inplace=True)
    
csv_list[0].head()
```

```python
sales_aggregated = pd.concat(csv_list, axis=1)

sales_aggregated.head()
# for col in sales_aggregated.columns:
#     print(col, sep=", ", end=", ")
```

```python
#some values are not rounded so we run this
sales_aggregated = sales_aggregated.applymap(lambda x: round(x, 3))
```

```python
sales_aggregated.info()

sales_aggregated.to_csv('./out/sales_aggregated.csv')
```
